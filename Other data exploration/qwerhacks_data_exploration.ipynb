```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Source: https://www.cdss.ca.gov/inforesources/research-and-data/calworks-data-tables/abcd350 
labels = ["Queer or another sexual orientation","Bisexual","Gay or Lesbian","Straight/Heterosexual"]
n_arr = np.array([1386,3522,3471,133682]) # Correspond with values in the labels list
n_tot = np.sum(n_arr) # total # of people that reported their sexual orientation 
n_per = n_arr/n_tot  # calculates a group's percentage of the total of

fig,ax = plt.subplots(1)
ax.pie(n_per,labels = labels,autopct='%1.1f%%')
ax.set(title = "Non-Assistance CalFresh beneficiaries by sexual orientation in 2021",
       xlabel="Data: ABCD 350 - Annual Recipient Report (2021)")

plt.tight_layout()
```
